```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import numpy as np
from matplotlib.patches import Circle, Wedge, Polygon
from matplotlib.collections import PatchCollection
```

```python
# First example
vertices = [(     0.875,      0.875,          1 ),
   (     0.875,   0.984375,      0.875 ),
   (         0,          1,          1 )]
```

```python
x = []
y = []
z = []
for vertex in vertices:
    p1, p2, p3 = vertex
    x += [p1] ; y += [p2] ;z += [p3]
print(x, y, z)
```

```python
# First example

plt.style.use('seaborn-paper')
#plt.style.use('default')

plt.rc('text', usetex=True)
plt.rc('text.latex', preamble=r'\usepackage{amsmath},\usepackage{amssymb}')
plt.rc('font', family='serif')

fig = plt.figure(figsize=(8, 6), dpi=300)
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, marker='o', c="g")
ax.plot3D(x, y, z, c="g")

ax.set_xbound(lower=0, upper=1.2)
ax.set_ybound(lower=0, upper=1.2)
ax.set_zbound(lower=0, upper=1.2)
ax.set_xlabel(r"$\mathbb{P}^\sigma_s(\diamondsuit_{1:\leq 4} \, \{s_3\})$")
ax.set_ylabel(r"$\mathbb{P}^\sigma_s(\diamondsuit_{1:\leq 8} \, \{s_3\})$")
ax.set_zlabel(r"$\mathbb{P}^\sigma_s(\diamondsuit_{2:\leq 700} \, \{s_3\})$")

ax.view_init(elev=20, azim=35.)
ax.invert_xaxis()
#ax.view_init(elev=30, azim=-100.)
plt.show()
```

```python
plt.style.use('seaborn-paper')
plt.rc('text', usetex=True)
plt.rc('text.latex', preamble=r'\usepackage{amsmath},\usepackage{amssymb}')
plt.rc('font', family='serif')

fig = plt.figure(figsize=(8, 6), dpi=300)
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, marker='o', c="g")
#ax.plot3D(x, y, z, c="g")

ax.set_xbound(lower=0, upper=1.2)
ax.set_ybound(lower=0, upper=1.2)
ax.set_zbound(lower=0, upper=1.2)
ax.set_xlabel(r"$\mathbb{P}^\sigma_s(\diamondsuit_{1:\leq 4} \, \{s_3\})$")
ax.set_ylabel(r"$\mathbb{P}^\sigma_s(\diamondsuit_{1:\leq 8} \, \{s_3\})$")
ax.set_zlabel(r"$\mathbb{P}^\sigma_s(\diamondsuit_{2:\leq 700} \, \{s_3\})$")

ax.add_collection3d(Poly3DCollection([vertices], facecolors='g'))

ax.view_init(elev=20, azim=45.)
ax.invert_xaxis()
plt.show()
```

```python
# Second example
xmax = [0.875]
ymax = [1]

xmin = [0.875, 0]
ymin = [0.875, 1]

plt.style.use('seaborn-poster')

plt.rc('text', usetex=True)
plt.rc('text.latex', preamble=r'\usepackage{amsmath},\usepackage{amssymb}')
plt.rc('font', family='serif')

fig = plt.figure(figsize=(6, 6), dpi=300)
ax = fig.add_subplot(111)

ax.set_xlim(-0.05, 1.05)
ax.set_ylim(-0.05, 1.05)
#ax.set_ylim(0.55, 1.15)

patches = [Polygon(list(zip(xmax + xmin, ymax + ymin)))]
p = PatchCollection(patches, alpha=0.2)
#p.set_array(np.array(colors))
ax.add_collection(p)

ax.scatter(xmax, ymax, marker='o', c="g", s=25)
ax.plot(xmin, ymin, c='#A80039', ls='--', lw=1.5)
ax.scatter(xmin, ymin, marker='o', c='#A80039', s=25)

ax.set_xlabel(r"$\mathbb{P}^\sigma_s(\diamondsuit_{1:\leq 4} \, \{s_3\})$")
ax.set_ylabel(r"$\mathbb{P}^\sigma_s(\diamondsuit_{2:\leq 700} \, \{s_3\})$")
#font = {'size': 12}
#plt.text(0.75, 0.93, r'$U_{\mathcal{Q}}$', fontdict=font)

plt.tight_layout()

plt.show()
```
